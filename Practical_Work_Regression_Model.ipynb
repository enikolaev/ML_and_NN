# Модель регрессии


```python
import numpy as np
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
```

##  Учебная задача для модели линейной регрессии

### Создание датасета для регрессии

Создадим синтетический набор данных для изучения принципов функционирования регрессионной модели. Выполним также визуализацию данных

```python
plt.style.use('ggplot')

n_points=100
X = np.linspace(-5, 5, n_points)
y = 10 * X - 7

X_train = X[0::2].reshape(-1, 1)
y_train = y[0::2] + np.random.randn(int(n_points/2)) * 10

X_test = X[1::2].reshape(-1, 1)
y_test = y[1::2] + np.random.randn(int(n_points/2)) * 10

#визуализация данных
plt.figure(figsize=(10, 5))
plt.plot(X, y, label='real')
plt.scatter(X_train, y_train, label='train')
plt.scatter(X_test, y_test, label='test')
plt.legend(loc='best')
plt.show()
```

### Линейная регрессия

Создадим модель линейной регрессии и выполним ее обучение на обучающей выборке `(X_train, y_train)`, затем выполним предсказание для тестовых данных `(X_test, y_test)`

```python
from sklearn import linear_model
# Создание модели и ее обучение
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)

# Печатаем коэффициенты линейной регрессии
w=reg.coef_
print('Коэффициент w = ', w)
b=reg.intercept_
print('Коэффициент b = ', b)

# Предсказание для одной точки
x_test = 3
result=reg.predict([[x_test]])
print('Предсказанное значение для %d равно %d' % (x_test, result))

# Визуализация тестовой и обучающей выборок, а также прямых: исходной и предсказанной
# Выборки
plt.figure(figsize=(10, 5))
plt.scatter(X_train, y_train, label='train')
plt.scatter(X_test, y_test, label='test')

# Прямые
plt.plot(X, y, label='real', color='yellow')
plt.plot(X[1::2], X[1::2].reshape(-1, 1).dot(w) + b, label='predicted', color='green')

plt.legend()
plt.show()
```
